# Reading and Coversion to Images

```python
ConvertNeeded = False
```

```python
if (ConvertNeeded):
    # Reading all the parquet files
    p0 = pd.read_parquet('/kaggle/input/bengaliai-cv19/train_image_data_0.parquet',)
    p1 = pd.read_parquet('/kaggle/input/bengaliai-cv19/train_image_data_1.parquet',)
    p2 = pd.read_parquet('/kaggle/input/bengaliai-cv19/train_image_data_2.parquet',)
    #p3 = pd.read_parquet('/kaggle/input/bengaliai-cv19/train_image_data_3.parquet',)
```

```python
if (ConvertNeeded):
    # Reading train and class mappings csv files
    ptrain = pd.read_csv('/kaggle/input/bengaliai-cv19/train.csv')
    map1 = pd.read_csv('/kaggle/input/bengaliai-cv19/class_map.csv')
```

```python
if (ConvertNeeded):
    # Concat all image files got from parquet files into 1 var pdata
    pdata = pd.concat([p0,p1,p2],)
```

```python
if (ConvertNeeded):
    # Reshape 1 image file in p1 to (137, 236) for Checking
    n1 = p1.iloc[0,1:].values.astype(np.uint8).reshape(137,236)
    # Display
    py.imshow(n1,)
```

```python
if (ConvertNeeded):
    # Delete unnecessary vars to clear up memory
    del p0
    del p1
    del p2
    del n1
```

```python
if (ConvertNeeded):
    # Check print head
    pdata.head()
```

```python
if (ConvertNeeded):
    # Convert entire images data to 1 parquet file 
    pdata.to_parquet('train.parquet',index=False)
```

```python
if (ConvertNeeded):
    # Reshape all images in pdata to (137, 236) and store in all_images
    all_images = []

    for i in tqdm(range(pdata.shape[0])):
        all_images.append(pdata.iloc[i,1:].values.astype(np.uint8).reshape(137,236))
```

```python
if (ConvertNeeded):
    # Init X as a pandas dataframe
    X = pd.DataFrame()

    # Store image_id data in X
    X['image_id'] = pdata['image_id']

    # Add this images data to X
    X['images'] = all_images
```

```python
if (ConvertNeeded):
    # Pickle the X Dataframe
    import pickle

    with open('ImagesDataframe.pickle', 'wb') as f:
        pickle.dump(X, f)
```

```python
if (ConvertNeeded):
    # Save Images Parameters
    imagesFolder = "TrainImages"
    imagesNamePrefix = ""
```

```python
if (ConvertNeeded):
    # Store all the images in imagesFolder
    for i in tqdm(range(len(all_images))):
        Image.fromarray(all_images[i]).save(imagesFolder + "/" + imagesNamePrefix + str(i) + ".jpeg")
```

```python
if (ConvertNeeded):
    # Zip the imagesFolder
    import zipfile

    folderName = imagesFolder
    zipf = zipfile.ZipFile(folderName + ".zip", 'w', zipfile.ZIP_DEFLATED)# zipf is zipfile handle
    print("Zipping " + folderName + "...")
    for root, dirs, files in os.walk(folderName):
        for file in files:
            zipf.write(os.path.join(root, file))
            # print("Zipped " + file)
    zipf.close()
    print("Zipping Done")
```
